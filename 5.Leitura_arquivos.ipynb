```python
import pandas as pd
import numpy as np
```

## CSV

### CSV Input

```python
caminho_arquivo = 'files/annual-co2-emissions-per-country.csv'
df = pd.read_csv(caminho_arquivo)
df
```

### Vamos supor que queira apenas EUA e a partir do ano de 1980

```python
# Filtrando para Code == 'USA' e anos a partir de 1980
df = df[(df['Code'] == 'USA') & (df['Year'] >= 1980)]
df.head()
```

### Salvando em outro arquivo csv ... 

```python
df.to_csv('files/USA.csv',index=False)
```

## Excel

Pandas podem ler e escrever arquivos do Excel, tenha em mente, isso só importa dados. Não fórmulas nem imagens, lembrando que imagens ou macros podem bugar o método.

```python
file = 'files/Histórico Membros FinancEEL.xlsx'
df = pd.read_excel(file, nrows=61)
df.head()
```

```python
df.drop(['Participava de Entidade', 'Score de Entidades', 'Idade'] , axis=1, inplace=True)
df.head()
```

```python
df['Score_class'] = df['Score de Permanência'].apply(lambda x: 0 if x < 12 else 1)
df.head()
```

```python
df.to_excel('files/Excel_tratado.xlsx')
```
